```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_bus_1=pd.read_csv(r'C:\temp\df1.csv')
df_bus_2=pd.read_csv(r'C:\temp\df2.csv')
df_bus_3=pd.read_csv(r'C:\temp\df3.csv')
df_bus_4=pd.read_csv(r'C:\temp\df4.csv')
df_bus_5=pd.read_csv(r'C:\temp\df5.csv')
df_bus_6=pd.read_csv(r'C:\temp\df6.csv')
df_bus_7=pd.read_csv(r'C:\temp\df7.csv')
df_bus_8=pd.read_csv(r'C:\temp\df8.csv')
df_bus_9=pd.read_csv(r'C:\temp\df9.csv')
df_bus_10=pd.read_csv(r'C:\temp\df10.csv')

Final_df=pd.concat([df_bus_1,df_bus_2,df_bus_3,df_bus_4,df_bus_5,df_bus_6,
                    df_bus_7,df_bus_8,df_bus_9,df_bus_10],ignore_index=True)
Final_df
```

```python
#data cleaning
# data about the data
Final_df.info()
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0) 
```

```python
print(Final_df['Ratings'].apply(type))
```

```python
Final_df['Ratings'] = Final_df['Ratings'].apply(str)
Final_df['Ratings'] = Final_df['Ratings'].str.replace('\n', ' ', regex=False)
# Example of cleaning the data
Final_df['Ratings'] = pd.to_numeric(Final_df['Ratings'], errors='coerce').fillna(0).clip(lower=0, upper=5)  # Assuming ratings are between 0 and 5
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
#Dropping column 'Unnamed'
print(Final_df)
Final_df = Final_df.drop(columns=['Unnamed: 0'])
```

```python
print(Final_df)
```

```python
Final_df.dropna(subset=['Bus_type'], inplace=True)
```

```python
# Example of cleaning the data
Final_df['Seats_Available'] = pd.to_numeric(Final_df['Seats_Available'], errors='coerce').fillna(0).astype(int)
```

```python
# change dataframe to csv
Final_df.to_csv(r'C:\temp\Final_df.csv', index=False)
```
